Analysis with control for distribution type as well as data set complexity level

```python
# ==============================================================================
# ANALYSIS METHOD 1: Complexity-Stratified Comparison
# Analyze by complexity level: Low SP, Mid SP, High SP
# ==============================================================================


def categorize_complexity(filename):
    """Categorize dataset complexity based on filename"""
    if "high SP" in filename:
        return "High"
    elif "mid SP" in filename:
        return "Mid"
    elif "low SP" in filename:
        return "Low"
    else:
        return "Unknown"


def analyze_by_complexity(paths, distribution_name):
    """Analyze SD completion times stratified by complexity"""
    complexity_data = {"Low": [], "Mid": [], "High": []}
    
    for path in paths:
        filename = os.path.basename(path)
        complexity = categorize_complexity(filename)
        
        if complexity == "Unknown":
            continue
            
        df = pd.read_csv(path)
        
        # Calculate mean difference for each method vs simulation mean
        means_comparison = {
            "PERT SD": (df["PERT SD"].mean() - df[reference_column].mean()),
            "BB/2 SD": (df["BB/2 SD"].mean() - df[reference_column].mean()),
            "Lognormal SD": (df["Lognormal SD"].mean() - df[reference_column].mean()),
        }
        
        complexity_data[complexity].append({
            "filename": filename,
            "means": means_comparison,
            "sim_SD": df[reference_column].mean()
        })
    
    # Aggregate by complexity level
    results = {}
    for complexity_level in ["Low", "Mid", "High"]:
        if complexity_data[complexity_level]:
            results[complexity_level] = {
                method: sum([d["means"][method] for d in complexity_data[complexity_level]]) / len(complexity_data[complexity_level])
                for method in target_columns
            }
    
    return results


print("=" * 80)
print("METHOD 1: COMPLEXITY-STRATIFIED ANALYSIS")
print("=" * 80)


tri_complexity = analyze_by_complexity(triangular_paths, "Triangular")
log_complexity = analyze_by_complexity(Lognormal_paths, "Lognormal")
beta_complexity = analyze_by_complexity(Beta_Paths, "Beta")


print("\nTriangular Distribution - Mean Difference from Sim SD (by Complexity):")
print("-" * 80)
for complexity, results in tri_complexity.items():
    print(f"\n{complexity} Complexity:")
    for method, diff in results.items():
        print(f"  {method:20s}: {diff:8.2f} days")


print("\n\nLognormal Distribution - Mean Difference from Sim SD (by Complexity):")
print("-" * 80)
for complexity, results in log_complexity.items():
    print(f"\n{complexity} Complexity:")
    for method, diff in results.items():
        print(f"  {method:20s}: {diff:8.2f} days")


print("\n\nBeta Distribution - Mean Difference from Sim SD (by Complexity):")
print("-" * 80)
for complexity, results in beta_complexity.items():
    print(f"\n{complexity} Complexity:")
    for method, diff in results.items():
        print(f"  {method:20s}: {diff:8.2f} days")
```

```python
# ==============================================================================
# ANALYSIS METHOD 2: Percent Error by Complexity
# Measure percent error for each method stratified by complexity
# ==============================================================================


def percent_error_by_complexity(paths, distribution_name):
    """Calculate percent error stratified by complexity"""
    complexity_data = {"Low": [], "Mid": [], "High": []}
    
    for path in paths:
        filename = os.path.basename(path)
        complexity = categorize_complexity(filename)
        
        if complexity == "Unknown":
            continue
            
        df = pd.read_csv(path)
        
        # Calculate percent error for each method
        percent_errors = {
            "PERT SD": (((df["PERT SD"] - df[reference_column]).abs() / df[reference_column]) * 100).mean(),
            "BB/2 SD": (((df["BB/2 SD"] - df[reference_column]).abs() / df[reference_column]) * 100).mean(),
            "Lognormal SD": (((df["Lognormal SD"] - df[reference_column]).abs() / df[reference_column]) * 100).mean(),
        }
        
        complexity_data[complexity].append(percent_errors)
    
    # Aggregate by complexity level
    results = {}
    for complexity_level in ["Low", "Mid", "High"]:
        if complexity_data[complexity_level]:
            results[complexity_level] = {
                method: sum([d[method] for d in complexity_data[complexity_level]]) / len(complexity_data[complexity_level])
                for method in target_columns
            }
    
    return results


print("\n\n" + "=" * 80)
print("METHOD 2: PERCENT ERROR BY COMPLEXITY")
print("=" * 80)


tri_pe = percent_error_by_complexity(triangular_paths, "Triangular")
log_pe = percent_error_by_complexity(Lognormal_paths, "Lognormal")
beta_pe = percent_error_by_complexity(Beta_Paths, "Beta")


print("\nTriangular Distribution - Mean Percent Error (by Complexity):")
print("-" * 80)
for complexity, results in tri_pe.items():
    print(f"\n{complexity} Complexity:")
    for method, error in results.items():
        print(f"  {method:20s}: {error:6.2f}%")


print("\n\nLognormal Distribution - Mean Percent Error (by Complexity):")
print("-" * 80)
for complexity, results in log_pe.items():
    print(f"\n{complexity} Complexity:")
    for method, error in results.items():
        print(f"  {method:20s}: {error:6.2f}%")


print("\n\nBeta Distribution - Mean Percent Error (by Complexity):")
print("-" * 80)
for complexity, results in beta_pe.items():
    print(f"\n{complexity} Complexity:")
    for method, error in results.items():
        print(f"  {method:20s}: {error:6.2f}%")
```

Visualizations of complexity controlled analysis

```python
# ==============================================================================
# ANALYSIS METHOD 3: Visual Comparison - Complexity vs Percent Error
# Creates visualizations to show how complexity affects method accuracy
# ==============================================================================

fig, axes = plt.subplots(2, 3, figsize=(16, 10))
fig.suptitle('Method Accuracy by Dataset Complexity', fontsize=16, fontweight='bold')

distributions = [
    ("Triangular", tri_pe),
    ("Lognormal", log_pe),
    ("Beta", beta_pe)
]

for idx, (dist_name, pe_data) in enumerate(distributions):
    # Create dataframe for easier plotting
    pe_df = pd.DataFrame(pe_data).T
    
    # Plot 1: Bar chart for this distribution
    ax = axes[0, idx]
    pe_df.plot(kind='bar', ax=ax, legend=True)
    ax.set_title(f'{dist_name} Distribution')
    ax.set_ylabel('Mean Percent Error (%)')
    ax.set_xlabel('Complexity Level')
    ax.tick_params(axis='x', rotation=0)
    ax.legend(title='Method', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
    
    # Plot 2: Line chart showing trends
    ax = axes[1, idx]
    for method in target_columns:
        if method in pe_df.columns:
            complexities = [k for k in pe_data.keys()]
            values = [pe_data[k].get(method, 0) for k in complexities]
            ax.plot(complexities, values, marker='o', label=method)
    
    ax.set_title(f'{dist_name} Distribution - Error Trend')
    ax.set_ylabel('Mean Percent Error (%)')
    ax.set_xlabel('Complexity Level')
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nVisualizations created showing method performance across complexity levels.")
```

```python
# ==============================================================================
# ANALYSIS METHOD 4: Performance Ranking by Complexity
# Rank methods from best to worst for each complexity level
# ==============================================================================

def rank_methods_by_complexity(percent_error_dict):
    """Rank estimation methods by accuracy (lowest error = best)"""
    rankings = {}
    
    for complexity, methods_pe in percent_error_dict.items():
        # Sort by error (ascending = best first)
        sorted_methods = sorted(methods_pe.items(), key=lambda x: x[1])
        rankings[complexity] = sorted_methods
    
    return rankings

print("\n\n" + "=" * 80)
print("METHOD 4: PERFORMANCE RANKING BY COMPLEXITY")
print("=" * 80)

tri_ranking = rank_methods_by_complexity(tri_pe)
log_ranking = rank_methods_by_complexity(log_pe)
beta_ranking = rank_methods_by_complexity(beta_pe)

print("\nTriangular Distribution - Method Rankings (Best  Worst):")
print("-" * 80)
for complexity in ["Low", "Mid", "High"]:
    if complexity in tri_ranking:
        print(f"\n{complexity} Complexity:")
        for rank, (method, error) in enumerate(tri_ranking[complexity], 1):
            print(f"  {rank}. {method:20s}: {error:6.2f}% error")

print("\n\nLognormal Distribution - Method Rankings (Best  Worst):")
print("-" * 80)
for complexity in ["Low", "Mid", "High"]:
    if complexity in log_ranking:
        print(f"\n{complexity} Complexity:")
        for rank, (method, error) in enumerate(log_ranking[complexity], 1):
            print(f"  {rank}. {method:20s}: {error:6.2f}% error")

print("\n\nBeta Distribution - Method Rankings (Best  Worst):")
print("-" * 80)
for complexity in ["Low", "Mid", "High"]:
    if complexity in beta_ranking:
        print(f"\n{complexity} Complexity:")
        for rank, (method, error) in enumerate(beta_ranking[complexity], 1):
            print(f"  {rank}. {method:20s}: {error:6.2f}% error")
```

```python
# ==============================================================================
# ANALYSIS METHOD 5: Absolute Mean Completion Time by Complexity (DEBUGGED)
# Compare actual mean completion times for each method vs simulation
# ==============================================================================

def analyze_mean_values_by_complexity(paths):
    """Analyze actual mean completion times stratified by complexity"""
    complexities = ["Low", "Mid", "High"]
    all_columns = target_columns + [reference_column]
    complexity_means = {c: {col: None for col in all_columns} for c in complexities}
    complexity_counts = {c: 0 for c in complexities}
    # Sum means for averaging
    complexity_sums = {c: {col: 0 for col in all_columns} for c in complexities}

    for path in paths:
        filename = os.path.basename(path)
        complexity = categorize_complexity(filename)
        if complexity not in complexities:
            continue
        df = pd.read_csv(path)
        complexity_counts[complexity] += 1
        for col in all_columns:
            if col in df.columns:
                complexity_sums[complexity][col] += df[col].mean()

    # Compute averages
    for c in complexities:
        if complexity_counts[c] > 0:
            for col in all_columns:
                complexity_means[c][col] = complexity_sums[c][col] / complexity_counts[c]
        # else: leave as None

    return complexity_means

def print_mean_table(dist_name, means_dict):
    print(f"\n{dist_name} Distribution - Mean Completion Times (days):")
    print("-" * 80)
    for complexity in ["Low", "Mid", "High"]:
        print(f"\n{complexity} Complexity:")
        for col in [reference_column] + target_columns:
            val = means_dict[complexity].get(col, None)
            if val is not None:
                print(f"  {col:20s}: {val:8.2f}")
            else:
                print(f"  {col:20s}: {'N/A':>8}")

print("\n\n" + "=" * 80)
print("METHOD 5: MEAN COMPLETION TIME ANALYSIS BY COMPLEXITY (DEBUGGED)")
print("=" * 80)

tri_means = analyze_mean_values_by_complexity(triangular_paths)
log_means = analyze_mean_values_by_complexity(Lognormal_paths)
beta_means = analyze_mean_values_by_complexity(Beta_Paths)

print_mean_table("Triangular", tri_means)
print_mean_table("Lognormal", log_means)
print_mean_table("Beta", beta_means)
```
